```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# use seaborn plotting defaults
import seaborn as sns; sns.set()

from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score, cross_val_predict
from sklearn.tree import DecisionTreeClassifier 
from sklearn import metrics
```

```python
df = pd.read_csv("Bank_CreditScoring.csv")
df.head()
```

```python
cleanup_nums = {"Employment_Type":     {"Self_Employed": 1, 
                                        "employee": 2,
                                        "Fresh_Graduate": 3, 
                                        "government": 4,
                                        "employer": 5},
               "More_Than_One_Products":     {"yes": 1, 
                                              "no": 2},
               "Property_Type":     {"condominium": 1, 
                                     "terrace": 2,
                                     "flat": 3, 
                                     "bungalow": 4},
               "State":     {"Kuala Lumpur": 1, 
                             "Johor": 2,
                             "Selangor": 3, 
                             "Penang": 4,
                             "P.Pinang": 5,
                             "Pulau Penang": 6, 
                             "Johor B": 7,
                             "K.L": 8, 
                             "N.Sembilan": 9,
                             "N.S": 10,
                             "Sarawak": 11,
                             "Sabah": 12, 
                             "SWK": 13,
                             "Trengganu": 14,
                             "Kedah": 15},
               "Decision":     {"Accept": 1, 
                                "Reject": 0}
               }

df = df.replace(cleanup_nums)
df.head()
```

```python

X = df.drop("Decision", axis = 1)    # big X represents independent variable
y = df.Decision 

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1) # 70% training and 30% test
```

```python
clf = DecisionTreeClassifier()
clf = clf.fit(X_train, y_train)

#Predict the response for test dataset
y_pred = clf.predict(X_test)
y_pred
```

```python
print("Accuracy: ", metrics.accuracy_score(y_test, y_pred))
```

```python
from sklearn.cluster import KMeans 
from sklearn import datasets
```

```python
km = KMeans(n_clusters = 5, random_state=2)
km.fit(X)

df.Decision = df['Decision'].apply(str)
df_new = df.copy()
df_new = df_new.drop("Decision", axis=1)
df_new['decision']=km.labels_
df_new.head()
```
